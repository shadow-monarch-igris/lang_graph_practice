# simple sequential workflow for langgraph

```python
from langgraph.graph import StateGraph ,START,END
from typing import TypedDict
```

```python
### state define ###
class simpel_bmi(TypedDict):
    height: float
    weight: float
    bmi: float
```

```python
def cal_bmi(state:simpel_bmi) -> simpel_bmi:
    height = state["height"]
    weight = state["weight"]
    bmi = weight / (height ** 2)
    state["bmi"] = bmi
    return state
```

```python
###### Work flow Definition ######

# define graph 
graph = StateGraph(simpel_bmi)
# add node to your graph
graph.add_node("bmi_calculation",cal_bmi)
# add edges to your graph
graph.add_edge(START,"bmi_calculation")
graph.add_edge("bmi_calculation",END)
# compile the graph 
workflow = graph.compile()
```

```python
# execute the graph
initial_state:simpel_bmi = {
    "height": 1.75,
    "weight": 70.0, 
}   
final_state = workflow.invoke(initial_state)
print(final_state)
```

```python
from langgraph.graph import StateGraph ,START,END
from typing import TypedDict
# creat states
class chatbot_state(TypedDict):
    user_input: str
    bot_response: str
```

```python
# create node functions
def chat_response(state:chatbot_state) -> chatbot_state:
    user_input =state["user_input"]
    prompt = f"The user said: {user_input}. Please respond appropriately."
    response = model.invoke(prompt)
    state["bot_response"] = response
    return state
```

```python
# define graph
graph = StateGraph(chatbot_state)
# create a graph node
graph.add_node("chat_response",chat_response)
# add edges to your graph
graph.add_edge(START,"chat_response")
graph.add_edge("chat_response",END)
# compile the graph
workflow = graph.compile()
# execute the graph
initial_state:chatbot_state = {
    "user_input": "Hello, can you tell me about yourself?",
    "bot_response": "",
}   
final_state = workflow.invoke(initial_state)
print(final_state)
```
